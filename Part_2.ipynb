```python
# To pluck out specific columns
df['Sexuality']
```

```python
# These single columns plucked out of dataframes are actually SERIES (another datatype in pandas, which is basically a fancy word for array but with a lot more functionality.)
type(df['Sexuality'])
```

```python
# To pluck out multiple columns (Singular columns constitute SERIES. Multiple constitute DATAFRAMES, like the one below.)
df[['Hobbyist','Country','Sexuality']]
```

```python
# To have a look at all the columns
df.columns
```

```python
# To pluck out specific rows/columns, we can use "loc" or "iloc"
df.iloc[0] # This gives us the 1st row
```

```python
# To pluck out multiple rows
df.iloc[[0,2,4,6,8]]
```

```python
# To pluck out a continuous set of rows
df.iloc[0:5]
```

```python
# To pluck out rows from a specific column(s) using 
df.iloc[0:5, 0:2] # NOTE that the second argument (i.e the column field is not inclusive) just like the first argument (i.e row field)
```

```python
# To pluck out rows from a specific column(s) using "loc"
df.loc[0:5, ["Hobbyist", "Student", "EdLevel"]] # NOTE that the second argument (i.e the column field is inclusive)
```
