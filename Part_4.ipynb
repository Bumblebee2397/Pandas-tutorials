# Part 4: Filtering - Using Conditionals to Filter Rows and Columns

```python
# Load data
import pandas as pd
path = "C:/Users/Dell/Desktop/ML Specialization/My Work - Post COVID/Pandas tutorials - Corey Schafer/developer_survey_2019/survey_results_public.csv"
df = pd.read_csv(path)
```

```python
# Take a look at the data
df.head(10)
```

```python
# We can check how many people code as a hobby using filters
filt = (df['Hobbyist'] == 'Yes')
```

```python
# On applying the filter we see that we only get the rows with a "Yes" for "Hobbyist"
df[filt]
```

```python
# We get the same results with df.loc as well
df.loc[filt] # When we pass filt as the first argument we are basically saying to pluck out all rows from the dataset that have a "Yes"
```

```python
# This can pluck out one/more than one columns subject to our filter condition
df.loc[filt, 'Country']
```

```python
# NOTE we can't use the keywords 'and' & 'or'. We have to use '&' and '|' respectively
filt_hobby_country = (df["Hobbyist"] == 'Yes') & (df["Country"] == 'India')
df.loc[filt_hobby_country]
```

```python
# Set option to display all columns in the data
pd.set_option('display.max_columns', 85)
```

```python
# Have a look at the data
df.head()
```

```python
# Constructing a filter with a salary floor and country restriction
filt_salary = (df['ConvertedComp'] > 70000) & filt_hobby_country
df.loc[filt_salary, ['WorkRemote', 'WorkWeekHrs']]
```

```python
# We also check within this limited set of people whether they know certain languages

filt_lang = filt_salary & df['LanguageWorkedWith'].str.contains('Java', na=False) & df['LanguageWorkedWith'].str.contains('Java', na=False)
df.loc[filt_lang]
```
