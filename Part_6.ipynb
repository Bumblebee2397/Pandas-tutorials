# Part 6: Add/Remove Rows and Columns From DataFrames

```python
# Load data
import pandas as pd
path = "C:/Users/Dell/Desktop/ML Specialization/My Work - Post COVID/Pandas tutorials - Corey Schafer/developer_survey_2019/survey_results_public.csv"
df = pd.read_csv(path)
df = df.applymap(str)
```

```python
# Take a look at the data
df.head(10)
```

```python
# Append a column to the dataframe which is constituted of 3 other columns
df['Appended col'] = df['Respondent'] + ', ' + df['Hobbyist'] + ', ' + df['Country']
```

```python
# Have a look at the data
df.head()
```

```python
# We can also remove columns from the dataset using the "df.drop()"" method
df.drop(columns = ['Appended col'])
```

```python
# Append a new column
df['Attribute'] = df['Hobbyist'] + ', ' + df['OpenSourcer'] + ', ' + df['Student']
```

```python
# Have a look at the data
df.head()
```

```python
# Split the data in the "Attribute" column to different columns
df['Attribute'].str.split(',', expand = True) # We can add it to the dataset using df[['col1', 'col2']] = df[].str.split()
```

```python
# Adding a new row
df = df.append({'Hobbyist': 'Wtf'}, ignore_index = True)
```

```python
# Removing a row
df.drop(index = 88883)
```

```python
# We can drop certain rows based on a filtering criteria as well. Eg: We want to drop all rows having 'Student = No'
filt = (df['Student'] == 'No')
df.drop(df[filt].index)
```
