# Part 7: Sorting Data

```python
# Load data
import pandas as pd
path = "C:/Users/Dell/Desktop/ML Specialization/My Work - Post COVID/Pandas tutorials - Corey Schafer/developer_survey_2019/survey_results_public.csv"
df = pd.read_csv(path)
```

```python
# Set option to have a look at all the rows
pd.set_option('display.max_columns', 85)

# Take a look at the data
df.head(10)
```

```python
# We rename the ConvertedComp column to Salary
df.rename(columns = {'ConvertedComp':'Salary'}, inplace = True)
```

```python
# Have a look at all the column headers
df.columns
```

```python
# Sorting the Salary series
df.sort_values(by = 'Salary', ascending = False, inplace = True)
```

```python
# Have a look at the data
df.head(50)
```

```python
"""
NOTE that df.sort_values(by = [column_1, column_2], ascending = [T/F, T/F]) first sorts the dataframe according to column_1 and
its respective condition (i.e True/False). If there are identical elements in column_1, then we can sort the rows with identical
entries for column_1 according to the condition on columnn_2.
"""
df.sort_values(by = ['Salary', 'Country'], ascending = [False, True], inplace = True)
```

```python
# Have a look at the data
df.head(10)
```

```python
# Have a look at the data
df[['Country', 'Salary']].head(100)
```

```python
# We can also get the n largest values in any series
df['Salary'].nlargest(20)
```

```python
# To see only those rows in the dataframe that have 10 highest salaries
df.nlargest(10, 'Salary')
```
