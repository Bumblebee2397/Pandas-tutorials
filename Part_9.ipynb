# Part 9: Cleaning Data - Casting Datatypes and Handling Missing Values

```python
# Load data
import pandas as pd
path = "C:/Users/Dell/Desktop/ML Specialization/My Work - Post COVID/Pandas tutorials - Corey Schafer/developer_survey_2019/survey_results_public.csv"
na_vals = ['Missing', 'N/A', 'Not avail', 'Not available'] # We can create a list of values which are indicated as missing
df = pd.read_csv(path, na_values = na_vals) 
```

```python
# Set option to have a look at all the rows
pd.set_option('display.max_columns', 85)

# Take a look at the data
df.head(10)
```

```python
# Removing rows/columns with some/all missing values. 
# NOTE that "axis = index" drops rows and "axis = columns" drops columns.
# "how = any" drops any row having NaN or None entry in any of the columns. "how = all" drops only those rows which NaN or None entries in all the columns.
# subset = ['ConvertedComp',..] tells us which columns to look at while dropping rows
# Here setting "how = any" or "how = all" makes no difference as it is allowed to look at the "ConvertedComp" column only

df.dropna(axis = 'index', how = 'any', subset = ['ConvertedComp'], inplace = True)
```

```python
# To remove rows with custom values
# STEP 1: Replace say "missing/Not availabe/NA" with np.nan using df.replace('...', np.nan, inplace = True)
# STEP 2: Drop rows like done above using df.dropna(...)

# To fill missing values with a certain string/numerical data

df['UndergradMajor'].fillna('MISSING', inplace = True)
```

```python
# Check datatypes of indiviual series of our dataset
df.dtypes # objects basically mean strings
```

```python
# NOTE that NaN values (np.nan) are actually float values. So incase, we try to cast a column to int datatype in will throw an error as it can't convert NaN to int
df['Age'] = df['Age'].astype(int)
```

```python
# Years people have spent coding
df['YearsCode'].unique()
```

```python
# Editing anomalies
df['YearsCode'].replace('Less than 1 year', 0, inplace = True)
df['YearsCode'].replace('More than 50 years', 51, inplace = True)
```
