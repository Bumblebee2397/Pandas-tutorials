# Part 5: Updating Rows and Columns - Modifying Data Within DataFrames

```python
# Load data
import pandas as pd
path = "C:/Users/Dell/Desktop/ML Specialization/My Work - Post COVID/Pandas tutorials - Corey Schafer/developer_survey_2019/survey_results_public.csv"
df = pd.read_csv(path)
df = df.applymap(str)
```

```python
# Take a look at the data
df.head(10)
```

```python
# Have a look at all the columns of the dataset
df.columns
```

```python
# Change all the column headers to uppercase
df.columns = [x.upper() for x in df.columns]
```

```python
# Have a look at all the columns
df.columns
```

```python
# Have a look at the data
df.head()
```

```python
# Changing the column headers back to lower case
df.columns = [x.lower() for x in df.columns]
```

```python
# Have a look at the data
df.head()
```

```python
# We can also rename individual columns by using the df.rename() method and passing a dictionary as an argument
df.rename(columns={'hobbyist':'is coding a hobby?'}) # Set inplace = True for permanant change
```

```python
# Have a look at the data
df.head()
```

```python
# We can change values of certain row elements as well
df.loc[0:2, 'hobbyist'] = ['Yes', 'Yes', 'Yes']
```

```python
# Have a look at the data
df.head()
```

```python
# Altering the elements in sonewcontent to lowercase
df['sonewcontent'] = df['sonewcontent'].str.lower()
```

```python
# Have a look at the data
df.head()
```

```python
"""
Now we'll look at a more advanced way to edit and modify row and column elements:
1. apply - used to apply a specific function on a set of values. It can be used on a dataframe or a series object.
2. applymap - can apply a specific function on the each individual element of the dataframe. They work only on dataframes.
3. map - can apply a specific function only on series.
4. replace

Let's have a look at "APPLY"
"""

# Using apply
df['hobbyist'].apply(len)

#  Defining a function that we want to "apply()"
def change_case(x):
    return x.upper()

# Using the function object. NOTE that the fuction object has to be passed i.e without the braces.df
df['hobbyist'] = df['hobbyist'].apply(change_case)
```

```python
# Have a look at the data
df.head()
```

```python
# Until now we've seen the 'apply' method used on series. Now we'll see its usage on dataframes
df.apply(len) # We see that the 'apply' method when used gives the number of entries in each column
```

```python
# This gives us the minimum value in each series of the dataset (alphabetical or numerical)
df.apply(pd.Series.min)
```

```python
"""
NOTE: When we use "apply" method on a series, it is applied on all its elements. When we use "apply" method on a dataframe, it
is applied on all the series in the dataframe. We see above that len function when passed in the "apply" method gives the number
length (i.e no of elements) of each series.

Now we look at "APPLYMAP"
"""
df.applymap(len) # Gives the number of character of each element in the entire dataset
```

```python
# To make each element in the dataset lowercase
df.applymap(str.lower)
```

```python
"""
NOTE: The "applymap" method can be used only on dataframes. When we use "applymap" method, it is applied on all the elements in 
the dataframe.

Now we look at "MAP"
"""
df['undergradmajor'] = df['undergradmajor'].replace({'nan':'None'})
```

```python
# Have a look at the data
df.head()
```

```python
# Application of df[].replace() method
df['hobbyist'].replace({'YES':True, 'NO':False})
```

```python
# Application of df[].replace() method
df['hobbyist'].replace({True: 'YES', False: 'NO'})
```

```python
# Application of df[].map() method. NOTE that things which are not mentioned in the map are all automatically set to 'NaN', hence, BE CAREFUL!
df['hobbyist'] = df['hobbyist'].map({'YES':True, 'NO':False})
```
