# Part 3: Indexes - How to set, reset and use indexes

```python
# Load data
import pandas as pd
path = "C:/Users/Dell/Desktop/ML Specialization/My Work - Post COVID/Pandas tutorials - Corey Schafer/developer_survey_2019/survey_results_public.csv"
df = pd.read_csv(path)
```

```python
# Take a look at the data
df.head(10)
```

```python
# We can set indexes as per our convenience. Default indexes start from 0. Here we set the respondent number as our index.
df.set_index('Respondent', inplace = True)
```

```python
# To have a view of all 85 columns we set the option accordingly
pd.set_option("display.max_columns", 85)
```

```python
# Have a look at the data
df.head()
```

```python
# NOTE that we cannot use the default integers to access the rows when using "loc"
df.loc[0] # This does not give us the 1st row
```

```python
# NOTE that we cannot use the default integers to access the rows when using "loc"
df.loc[1] # This gives us the 1st row
```

```python
# "iloc" does not give an error
df.iloc[0] # This gives us the 1st row
```

```python
# Reset index
df.reset_index(inplace = True)
```

```python
# Have a look at data
df.head()
```
